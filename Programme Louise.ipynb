```python
import math;
def isPrime(n):
    if (math.floor(n) != n):
        print("Un nombre premier est forc√©ment un entier.")
        return False;
    else:
        if (n<=3):
            print(n, "est premier !");
            return True;
        else:
            i = 2;
            while i < n:
                if (n%i == 0):
                    print(n, "n'est PAS premier.");
                    return False;
                i = i + 1
            print(n, "est premier !");
            return True;
            
```

```python
isPrime(2003)
```

```python
test = range(0,900)
```

```python
res = [];
for i in test:
    res.append(isPrime(i));
```

```python
%matplotlib inline

import matplotlib;
import matplotlib.pyplot as pl;

point = [];
for i in res:
    #print(i);
    if (i == True):
        point.append(1);
    else:
        point.append(0);

point
pl.plot(point);

```
